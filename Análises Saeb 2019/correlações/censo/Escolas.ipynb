```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cols1 =['CO_ENTIDADE', 'IN_AGUA_POTAVEL', 'IN_ENERGIA_REDE_PUBLICA', 'IN_ESGOTO_REDE_PUBLICA', 'IN_LIXO_SERVICO_COLETA',
        'IN_BIBLIOTECA', 'IN_COZINHA', 'IN_PISCINA', 'IN_COMPUTADOR', 'QT_EQUIP_MULTIMIDIA', 'QT_SALAS_UTILIZA_CLIMATIZADAS',
      'QT_TABLET_ALUNO', 'IN_INTERNET','IN_LABORATORIO_CIENCIAS']



df = pd.read_csv('ESCOLAS.csv', usecols= cols1,  sep= '|', encoding= 'unicode_escape')
df.head()
```

```python
df = df.rename(columns={"CO_ENTIDADE": "ID_ESCOLA"})
```

```python
df.head()
```

```python
cols =['ID_ESCOLA', 'MEDIA_9EF_MT']

esc = pd.read_csv('TS_ESCOLA.csv', usecols= cols, encoding= 'unicode_escape')
esc.dropna(inplace=True)
esc.head()
```

```python
# Separando dados por niveis
escala = [
    (esc["MEDIA_9EF_MT"] < 200),
    (esc["MEDIA_9EF_MT"] >= 200) & (esc["MEDIA_9EF_MT"] < 225),
    (esc["MEDIA_9EF_MT"] >= 225) & (esc["MEDIA_9EF_MT"] < 250),
    (esc["MEDIA_9EF_MT"] >= 250) & (esc["MEDIA_9EF_MT"] < 275),
    (esc["MEDIA_9EF_MT"] >= 275) & (esc["MEDIA_9EF_MT"] < 300),
    (esc["MEDIA_9EF_MT"] >= 300) & (esc["MEDIA_9EF_MT"] < 325),
    (esc["MEDIA_9EF_MT"] >= 325) & (esc["MEDIA_9EF_MT"] < 350),
    (esc["MEDIA_9EF_MT"] >= 350) & (esc["MEDIA_9EF_MT"] < 375),
    (esc["MEDIA_9EF_MT"] >= 375) & (esc["MEDIA_9EF_MT"] < 400),
    (esc["MEDIA_9EF_MT"] >= 400)
    ]
nivel = ["Nivel 0", "Nivel 1", "Nivel 2", "Nivel 3", "Nivel 4", "Nivel 5", "Nivel 6", "Nivel 7", "Nivel 8", "Nivel 9"]
esc["ESCALA_MT"] = np.select(escala, nivel)
esc.head()
```

```python
esc.drop('MEDIA_9EF_MT', axis=1, inplace =True)
```

```python
esc.head()
```

```python
df.shape
```

```python
df_esc = pd.merge(df, esc, how = 'inner', on = 'ID_ESCOLA')
df_esc.head()
```

```python
df_aux= df_esc.iloc[:, 0:15]
df_aux.head()
```

```python
# Transformando dados categoricos
from sklearn import preprocessing

label = preprocessing.LabelEncoder()
data_encoded = pd.DataFrame() 

for i in df_aux.columns:
    data_encoded[i]=label.fit_transform(df_aux[i])
data_encoded.head()
```
