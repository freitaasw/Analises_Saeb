```python
import plotly
import plotly.express as px
import pandas as pd
import math
```

```python
# Lendo o arquivo
regiao = pd.read_excel("TS_REGIAO.xlsx")
regiao.head(20)
```

```python
# Verificando o tamanho 
regiao.shape
```

```python
# Verificando a existência de valores ausentes
regiao.isnull().sum()
```

```python
# Removendo todas as linhas que possuem valores ausentes (NaN)
regiao.dropna(axis=0, inplace=True)
```

```python
# Removendo valores duplicados
regiao.drop_duplicates(keep='first', inplace=True)
```

```python
regiao.head(20)
```

```python
regiao_agroup = regiao.groupby(["NO_REGIAO"]).mean()
regiao_agroup.head()
```

```python
regiao_agroup_LP = regiao_agroup['MEDIA_9_LP'].reset_index().sort_values(by='MEDIA_9_LP')
fig1 = px.bar(regiao_agroup_LP, x = 'NO_REGIAO', y = 'MEDIA_9_LP',
             labels={'NO_REGIAO': 'Regiões', 'MEDIA_9_LP': 'Notas'},
             color_discrete_sequence=px.colors.qualitative.T10, template='presentation', text='MEDIA_9_LP')
fig1.update_traces(textposition='inside',texttemplate='%{text:.5s}')
fig1.update_layout(title={'text' : 'Médias de Língua portuguesa, 9° ano em 2019', 'y': 0.95, 'x': 0.5})
fig1.show()
```

```python
dN = regiao[regiao["NO_REGIAO"] == "Nordeste"]
dN["MEDIA_9_LP"].describe()
```

```python
dS = regiao[regiao["NO_REGIAO"] == "Sul"]
dS["MEDIA_9_LP"].describe()
```

Teste T-Student para a Média das Notas de Português do Nordeste e Sul

Hipótese do Teste

* Hipótese Nula (H0): Não há diferença significativa entre as médias
* Hipótese Alternativa (H1): Há diferença significativa entre as médias



```python
# Calculando a variância
s_variancia = (((dN["MEDIA_9_LP"].count() - 1) * pow(dN["MEDIA_9_LP"].std(), 2)) + ((dS["MEDIA_9_LP"].count() - 1) * pow(dS["MEDIA_9_LP"].std(), 2))) / ((dN["MEDIA_9_LP"].count() + dS["MEDIA_9_LP"].count()) - 2)

s_variancia
```

```python
# Calculando o valor de t
t_calculado = (dN["MEDIA_9_LP"].mean() - dS["MEDIA_9_LP"].mean()) / math.sqrt(s_variancia * ((1/dN["MEDIA_9_LP"].count()) + (1/dS["MEDIA_9_LP"].count())))

t_calculado
```

Verificar na tabela estatística de t-student o valor crítico para t

Grau de Liberdade: gl = n1 + n2 - 2     =>      gl = 48 + 48 - 2 = 94

Significancia de 0.05 (95% de certeza)  =>      t_critico = 1.9867



Se t_calculado >= t_critico rejeita H0 e aceita H1

t_calculado = |-3.32| = 3.32

t_critico = |1.9867| = 1.9867

Conclusão: Como o t_calculado é maior que o t_critico aceita-se a hipótese H1. Dessa forma, conclui-se que existe diferença estatisticamente significativa entre as notas de português do Nordeste e Sul.
